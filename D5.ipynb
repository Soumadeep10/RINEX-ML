```python
overs = [0,4,8,12,16,20]
team1 = [0,30,67,98,123,145]
team2 = [0,27,75,103,110,135]
plt.plot(overs,team1,label='Team1')
plt.plot(overs,team2,label='Team2')
plt.title('Analysis of Cricket Match',fontsize=20)
plt.xlabel('OVERS')
plt.ylabel('RUNS')
plt.legend()
```

```python
#SPORTSWATCH DATA
x =np.array([60,70,80,90,100])
y = [250,270,260,280,300]
plt.plot(x,y)
plt.title('SPORTS WATCH DATA',fontsize = 30)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.grid()# will create grid lines
```

```python
#SUB PLOT
fig = plt.figure()

x1 = [1,2,3,4]
x2 = [5,6,7,8]
y1 = np.random.randint(0,10,4)
y2 = np.random.randint(0,10,4)

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(224)

ax1.plot(x1,y1)
ax2.plot(x2,y2)
```

```python
#SCATTER PLOT - POINTS
a = [1,2,3,4,5]
b = [2,3,4,5,6]
plt.scatter(a,b,marker = 'D',color = 'orange')
plt.plot(a,b)
```

```python
a = [1,2,3,4,5,6,7,8,9]
b = [5,4,6,2,3,1,7,5,8]
plt.scatter(a,b,color = 'green',marker = '*')
```

```python
#BAR GRAPH
name = ['A','B','C','D','E']
num = [4,2,7,5,1]
plt.bar(name,num,color = ['orangered','lime','orange','green','crimson'])
plt.title('Name vs Numbers',fontsize = 20)
plt.xlabel('Names',fontsize = 20)
plt.ylabel('Numbers',fontsize = 20)
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y,color = 'pink',width = 0.1)
```

```python
#Horizontal Bar Garph - we use 'plt.barh'
name = ['Raj','Thomas','Mac','Naresh']
weight = [58,67,60,71]
plt.barh(name,weight,color = ['orangered','lime','tomato','crimson'])
plt.title('NAME VS WEIGHT',fontsize = 20)
plt.xlabel('WEIGHT',fontsize=20)
plt.ylabel('NAME',fontsize=20)
```

```python
#PIECHART
sales = ['10','25','16','37']
brands = ['Maruti','Ford','Volvo','Suzuki']
color = ['lime','red','blue','crimson']
plt.pie(sales,labels = brands,colors = color,explode = [0.3,0,0,0])
```
